```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

```

```python
tmax = 100
dt = 1e-3
D = 1
def ou():
    t = 0 
    o = []
    x = 0
    while t<tmax:
        t+=dt
        x = x + np.random.normal()*np.sqrt(2*dt) - x*dt
        o.append(x)
    return o
u = ou()
t = np.arange(0,tmax,dt)
```

```python
plt.plot(t,u)
```
