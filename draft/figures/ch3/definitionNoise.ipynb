```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

```

```python
tmax = 100
dt = 1e-3
D = 1
np.random.seed(101)
def ou():
    t = 0 
    o = []
    x = 0
    while t<tmax:
        t+=dt
        x = x + np.random.normal()*np.sqrt(2*dt) - x*dt
        o.append(x)
    return o
u = ou()
t = np.arange(0,tmax,dt)
```

```python
# initialize the figure
h = 1.25 # height
w = 5.5 # width

fig, ax = plt.subplots(1,1,figsize=(w,h),dpi=1000,tight_layout=True)
plt.axis('off')
ax.plot(t,u,lw=0.5,color='black')

plt.savefig('noise.pdf',dpi=1000)
```
